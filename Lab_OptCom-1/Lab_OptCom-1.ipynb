```python
import numpy as np
import matplotlib.pyplot as plt

point    = np.linspace(0, 11, num=12)
data     = np.array([-7.53, -8.44, -9.36, -10.32, -11.22, -12.19,
                    -13.16, -14.11, -15.07, -16.07, -17.08, -18.07]) # Measured values
teorical = np.linspace(-6.92, -17.92, num=12) # Theoretical values

ins_loss = data[0] - teorical[0] # Insertion loss of VOA
error    = data-teorical

plt.figure(figsize=(10,6))
plt.plot(point, data, '-o', linewidth=2.0, label= 'Expected')
plt.plot(point, teorical, '-o', linewidth=2.0, label= 'Measured')
plt.title('Power measurement', fontsize=16.0, fontstyle="oblique")
plt.xlabel('Attenuation (dB)', fontsize=16.0, fontstyle="oblique")
plt.ylabel('Power (dBm)', fontsize=16.0, fontstyle="oblique")
plt.legend(loc="lower left", fontsize=12.0)
plt.grid()
plt.show()

plt.figure(figsize=(10,6))
plt.plot(point, error, '-o', linewidth=2.0)
plt.title('Measurement error', fontsize=16.0, fontstyle="oblique")
plt.xlabel('Attenuation (dB)', fontsize=16.0, fontstyle="oblique")
plt.ylabel('Error (dB)', fontsize=16.0, fontstyle="oblique")
plt.grid()
plt.show()
```
